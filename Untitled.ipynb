```python
# Import dependencies
import pandas as pd
import numpy as np
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt
import scipy.stats as sts
from scipy.stats import linregress
from scipy.stats import f_oneway
import gmaps
import random
#from config import gkey
import requests
import json
```

```python
random.seed()
n = []
for i in range(0,10000):
    n.append(random.normalvariate(0,1))

plt.hist(n)
```

```python
m = 0.5
b = 0
std =200
x=[]
y=[]
for i in range(0,1000+1):
    x.append(float(i))
    y_val = m*i+b+random.normalvariate(0,std)
    y.append(y_val)
```

```python
plt.scatter(x,y)
min_x = min(x)
max_y = max(y)
regress_values_N = []
(slope_N, intercept_N, rvalue_N, pvalue_N, stderr_N) = linregress(x,y)
line_eq = "y = " + str(round(slope_N,3)) + "x + " + str(round(intercept_N,3))+", r = " +str(round(rvalue_N,3))\
                        + " p = "+str(round(pvalue_N,3))
for i in range(0,len(x)):
    regress_values_N.append(x[i] * slope_N + intercept_N)
plt.plot(x,regress_values_N,"r-")
plt.annotate(line_eq,(min_x+1,max_y-1),fontsize=11,color="red") #writes to coordinates on graph
```

```python
x_min = min(x)
x_max = max(x)
delta_x = x_max-x_min
bin_size = 100 #number of years
df = pd.DataFrame()
bins = [(x_min+i*bin_size) for i in range(0,int(delta_x/bin_size)+1)]
groups = [((i)*bin_size+x_min) for i in range(1,int(delta_x/bin_size)+1)]
df["x"]=x
df["y"]=y
df["Bin"]=pd.cut(df["x"],bins, labels = groups, include_lowest = True)


#box plot of full data in bins
red_diamond = dict(markerfacecolor='r', marker='D')
fig1, ax1 = plt.subplots()
ax1.boxplot([df[(df["Bin"]==i)]["y"]for i in groups],labels=groups,showfliers = True,
             flierprops = red_diamond) 
plt.show()

bin = []
for i in range(0,len(groups)):
         bin.append(df[df["Bin"]==groups[i]]["y"].tolist())
s_list=[]
p_list=[]
for i in range(1,len(groups)):
    [s,p] = f_oneway(bin[i], bin[i-1])
    s_list.append(s)
    p_list.append(p)
    print(f"bins {groups[i-1]} and {groups[i]}: {f_oneway(bin[i], bin[i-1])}")
        
avg_p= sts.describe(p_list)[2]
var_p = sts.describe(p_list)[3]
max_p = sts.describe(p_list)[1][1]
max_p = sts.describe(p_list)[1][0]
```

```python
Summary_df = pd.DataFrame(columns = ['slope', "Stand_Dev Noise", "Regress m", "Regress b", "Regress r", "Regress P",\
                                    "Average P", "Var P", "Max P", "Min P" ])
for n in range(0,200,2):
    for std_dev in range(50,150,5):
        m=n/200
        b = 0
        std =std_dev
        x=[]
        y=[]
        for i in range(0,1000+1):
            x.append(float(i))
            y_val = m*i+b+random.normalvariate(0,std)
            y.append(y_val) 
        
        (slope, intercept, rvalue, pvalue, stderr) = linregress(x,y)
        x_min = min(x)
        x_max = max(x)
        delta_x = x_max-x_min
        bin_size = 100 #number of years
        df = pd.DataFrame()
        bins = [(x_min+i*bin_size) for i in range(0,int(delta_x/bin_size)+1)]
        groups = [((i)*bin_size+x_min) for i in range(1,int(delta_x/bin_size)+1)]
        df["x"]=x
        df["y"]=y
        df["Bin"]=pd.cut(df["x"],bins, labels = groups, include_lowest = True)


        #box plot of full data in bins
#             red_diamond = dict(markerfacecolor='r', marker='D')
#             fig1, ax1 = plt.subplots()
#             ax1.boxplot([df[(df["Bin"]==i)]["y"]for i in groups],labels=groups,showfliers = True,
#                          flierprops = red_diamond) 
#             plt.show()

        bin = []
        for i in range(0,len(groups)):
                 bin.append(df[df["Bin"]==groups[i]]["y"].tolist())
        s_list=[]
        p_list=[]
        for i in range(1,len(groups)):
            [s,p] = f_oneway(bin[i], bin[i-1])
            s_list.append(s)
            p_list.append(p)
            #print(f"bins {groups[i-1]} and {groups[i]}: {f_oneway(bin[i], bin[i-1])}")

        avg_p= sts.describe(p_list)[2]
        var_p = sts.describe(p_list)[3]
        max_p = sts.describe(p_list)[1][1]
        min_p = sts.describe(p_list)[1][0]
        
        Summary_df = Summary_df.append({"slope":m, "Stand_Dev Noise": std,"Regress m":slope,
                                      "Regress b":intercept,"Regress r":rvalue,"Regress P":pvalue,
                                      "Average P":avg_p,"Var P":var_p,
                                      "Max P":max_p, "Min P":min_p}, ignore_index = True)
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import numpy as np


fig = plt.figure()
ax = fig.gca(projection='3d')

# Make data.
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

# Plot the surface.
surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)

# Customize the z axis.
ax.set_zlim(-1.01, 1.01)
ax.zaxis.set_major_locator(LinearLocator(10))
ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))

# Add a color bar which maps values to colors.
fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
```
